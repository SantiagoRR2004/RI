<b>Lectura como operación prioritaria en Recuperación de Información</b><br><br>

En los sistemas de <b>Recuperación de Información (RI)</b> la operación más prioritaria es la <b>lectura</b>.  
El objetivo de un sistema RI es localizar y recuperar información relevante dentro de grandes volúmenes de datos.  
La eficiencia está centrada en acceder y filtrar documentos rápidamente, más que en modificarlos.

En cambio, en los <b>Sistemas de Gestión de Bases de Datos (SGBD)</b> actuales, las operaciones de <b>lectura, escritura, actualización y borrado</b> son igualmente importantes.  
Los motores de bases de datos buscan mantener un equilibrio entre rendimiento de consultas e integridad de los datos (ACID).  
Por eso, aunque permiten búsquedas, no están optimizados para búsquedas complejas en grandes volúmenes de datos no estructurados.

De ahí la necesidad de <b>aplicaciones dedicadas</b> como Lucene, Elasticsearch o Solr, que implementan índices y algoritmos especializados de búsqueda, sin las limitaciones de los SGBD tradicionales cuando los datos son muy voluminosos o no están estructurados.

<b>Modelo orientado a documentos</b><br><br>

El <b>modelo orientado a documentos</b> es un esquema de representación donde cada documento se trata como una unidad independiente.  
Podemos imaginarlo como una “tabla” dentro de una base de datos, pero con la diferencia de que cada fila (documento) es un conjunto de <b>campos flexibles</b> (pares <i>clave-valor</i>) que no requieren un esquema fijo.

Características principales:

<ul>
<li>Cada documento es independiente, similar a un registro en una tabla.</li>
<li>Los documentos no necesitan tener los mismos campos obligatoriamente.</li>
<li>El motor puede <b>evaluar y buscar</b> en los documentos usando cualquier campo indexado.</li>
<li>Esto lo hace muy adecuado para datos no estructurados o semiestructurados.</li>
</ul>

## Ejemplos de motores de búsqueda basados en índices invertidos

<ul>
<li><b>Lucene</b>: es la librería base más utilizada y está escrita en Java. Destaca por su <b>índice invertido</b>, que lo hace extremadamente <b>rápido</b> en búsquedas, optimizado para manejar documentos a gran escala y con capacidad de desbloqueo eficiente. Su fortaleza es la velocidad y la optimización del motor interno.</li><br>

<li><b>Solr</b>: utiliza la librería <b>Lucene</b> como núcleo, pero añade funcionalidades adicionales como una <b>API en distintos lenguajes</b>, herramientas de administración y mayor facilidad de integración. Se usa para aplicaciones empresariales donde se requieren consultas complejas y administración avanzada.</li><br>

<li><b>ElasticSearch</b>: construido sobre Lucene, ofrece una <b>API REST</b> muy extendida. Es la opción más popular actualmente porque permite trabajar en <b>tiempo real</b>, escalar en <b>clusters y nodos</b>, y es ampliamente usado en big data y analítica.</li><br>

<li><b>PyLucene</b>: Es un enlace (binding) de Python para Apache Lucene, que permite usar toda la potencia del motor de búsqueda de Lucene desde Python.</li><br>

<li><b>Whoosh</b>: es una librería de búsqueda en Python puro. Destaca por su <b>índice invertido</b>, lo que lo hace extremadamente <b>rápido</b> en búsquedas. Está optimizado para manejar documentos a gran escala y ofrece un <b>desempeño eficiente</b> incluso sin un motor externo. Su fortaleza principal es la velocidad y la simplicidad de implementación, siendo ideal para proyectos que requieren búsqueda de texto completo sin depender de Java o servicios externos.</li>
</ul>

## El índice invertido: el corazón del motor de búsqueda

El <b>índice invertido</b> es una estructura de datos en formato <b>clave-valor</b>:  
- La clave es un término del vocabulario.  
- El valor es la lista de documentos en los que aparece ese término.  
 

Su función es indicar rápidamente <b>en qué documentos aparece cada término</b>, permitiendo búsquedas muy eficientes.  
Constituye el verdadero <b>corazón de un motor de búsqueda</b>.

<h2 style="text-align:center; color:#004466; margin-bottom:20px; text-shadow:1px 1px 2px #ccc;">Arquitectura de Whoosh y ejemplo mínimo</h2>
<p style="font-size:15px; text-align:center; margin-bottom:20px;">Whoosh funciona en varias fases clave:</p>
<ul style="max-width:850px; margin:auto; font-size:14px; line-height:1.6;">
<li><b>Analyze document</b>: el texto de entrada se procesa con un <i>Analyzer</i> (tokenización, normalización, stopwords, etc.).</li>
<li><b>Index document</b>: se crean <b>Document</b> con <b>Field</b> (campos) que se añaden al índice.</li>
<li><b>Build query</b>: se construyen consultas (<i>Query</i>) desde texto plano o expresiones.</li>
<li><b>Run query</b>: el <i>Searcher</i> busca en el índice los documentos relevantes.</li>
<li><b>Render results</b>: se devuelven documentos y puntuaciones.</li>
</ul>
<hr style="margin:30px 0; border:0; border-top:2px dashed #bbb;">
<h3 style="text-align:center; color:#006633; margin-bottom:15px;">Componentes básicos</h3>
<ul style="max-width:900px; margin:auto; font-size:14px; line-height:1.7;">
<li><b>Esquema (Schema)</b>: Define la <b>estructura de los documentos</b> que se van a indexar.
<ul>
<li>Cada campo (por ejemplo <code>title</code> o <code>content</code>) tiene un tipo: <code>TEXT</code>, <code>ID</code>, <code>KEYWORD</code>, etc.</li>
<li>Es el equivalente a un “modelo” de documento en Whoosh.</li>
</ul>
</li>
<li><b>Fields</b>: son los <b>atributos</b> de cada campo del esquema del documento. Whoosh incluye tipos como:
<ul>
<li><code>ID</code>: identificador único, no analizado.</li>
<li><code>TEXT</code>: texto analizado (tokenizado), usado para búsquedas.</li>
<li><code>KEYWORD</code>: lista de términos separados por comas, útil para etiquetas.</li>
<li><code>NUMERIC</code>: enteros o decimales indexables.</li>
<li><code>DATETIME</code>: fecha y hora.</li>
</ul>
</li>
<li><b>Indexación (Indexing)</b>: procesa y añade documentos al índice mediante un <b>writer</b>. Los <b>writers</b> son los componentes encargados de modificar el índice: agregar, actualizar o eliminar documentos. Existen varios tipos:
<ul>
<li><code>IndexWriter</code> (básico):
<ul>
<li>Bloquea el índice mientras se escribe, para evitar corrupción de datos.</li>
<li>Permite añadir, actualizar o borrar documentos.</li>
<li>Es seguro, pero puede ser más lento si se hacen muchas operaciones pequeñas.</li>
</ul>
<pre style="background:#fff; border:1px solid #999; border-radius:6px; padding:10px;">writer = ix.writer()</pre>
</li>
<li><code>AsyncWriter</code>:
<ul>
<li>Diseñado para aplicaciones concurrentes.</li>
<li>Abre un writer temporal y luego fusiona los cambios al confirmar (<code>commit</code>).</li>
<li>Reduce bloqueos y es más eficiente en entornos multiusuario.</li>
</ul>
<pre style="background:#fff; border:1px solid #999; border-radius:6px; padding:10px;">from whoosh.writing import AsyncWriter
writer = AsyncWriter(ix)</pre>
</li>
<li><code>BufferedWriter</code>:
<ul>
<li>Mantiene en memoria las escrituras hasta un límite de documentos o un tiempo máximo.</li>
<li>Luego realiza la escritura real en el índice.</li>
<li>Es útil para mejorar el rendimiento cuando se agregan documentos en lotes grandes.</li>
</ul>
<pre style="background:#fff; border:1px solid #999; border-radius:6px; padding:10px;">from whoosh.writing import BufferedWriter
writer = BufferedWriter(ix, limit=100, period=60)</pre>
</li>
</ul>
<!--Ejemplo: el texto <code>"Hola Mundo Whoosh"</code> se tokeniza como <code>["hola", "mundo", "whoosh"]</code>.</li><br>-->
</li>
<li><b>Análisis de texto (Analyzer)</b>: define cómo se transforma el texto. Los analizadores procesan el texto de los documentos y las consultas: tokenización, minúsculas, eliminación de stopwords, stemming, etc. Es plugable, lo que significa que puedes personalizar cómo se interpreta el texto. Whoosh trae analizadores como:
<ul>
<li><code>StandardAnalyzer</code> (estándar): El más común. Divide el texto en tokens, pasa a minúsculas y elimina stopwords.
<pre style="display:block; text-align:left; border:1px solid #990000; border-radius:6px; padding:10px; background:#fff; margin:10px auto;">from whoosh.analysis import StandardAnalyzer
analyzer = StandardAnalyzer()</pre>
</li>
<li><code>SimpleAnalyzer</code>: Divide el texto en tokens por caracteres alfanuméricos, pasa todo a minúsculas y no elimina stopwords.
<pre style="display:block; text-align:left; border:1px solid #990000; border-radius:6px; padding:10px; background:#fff; margin:10px auto;">from whoosh.analysis import SimpleAnalyzer
analyzer = SimpleAnalyzer()</pre>
</li>
<li><code>KeywordAnalyzer</code>: Trata todo el texto como un solo token. Útil para campos que no deben dividirse.
<pre style="display:block; text-align:left; border:1px solid #990000; border-radius:6px; padding:10px; background:#fff; margin:10px auto;">from whoosh.analysis import KeywordAnalyzer
analyzer = KeywordAnalyzer()</pre>
</li>
<li><code>RegexAnalyzer</code>: Permite definir un patrón de expresión regular para tokenizar.
<pre style="display:block; text-align:left; border:1px solid #990000; border-radius:6px; padding:10px; background:#fff; margin:10px auto;">from whoosh.analysis import RegexAnalyzer
analyzer = RegexAnalyzer(r"\w+")</pre>
</li>
<li>Analizadores avanzados:
<ul>
<li><code>StemmingAnalyzer</code>: Aplica un algoritmo de stemming (reduce palabras a su raíz, ej. “corriendo” → “correr”).
<pre style="display:block; text-align:left; border:1px solid #990000; border-radius:6px; padding:10px; background:#fff; margin:10px auto;">from whoosh.analysis import StemmingAnalyzer
analyzer = StemmingAnalyzer()</pre>
</li>
<li>Analizadores para lenguajes específicos con <code>SnowballStemmer</code>:
<pre style="display:block; text-align:left; border:1px solid #990000; border-radius:6px; padding:10px; background:#fff; margin:10px auto;">from whoosh.analysis import StemmingAnalyzer
from whoosh.lang.snowball import SpanishStemmer
analyzer = StemmingAnalyzer(stemfn=SpanishStemmer())</pre>
</li>
</ul>
</li>
</ul>
</li>

<li><b>Consulta (Query)</b>: El parser convierte las consultas de texto en objetos <code>Query</code>.
<ul>
<li>Soporta operadores booleanos, búsqueda por proximidad, coincidencia aproximada y combinaciones complejas.</li>
<li>El motor de búsqueda ejecuta la consulta contra el índice invertido para obtener los documentos relevantes.</li>
</ul>
</li>

<li><b>Almacenamiento (Storage)</b>: sistema que utiliza Whoosh para guardar los índices en disco o en memoria.
<ul>
<li><code>FileStorage</code>: Guarda los índices en archivos en disco.
<ul>
<li>Se suele usar cuando quieres que los índices persistan entre sesiones.</li>
<li>Puedes especificar la carpeta donde se almacenarán los archivos.</li>
</ul>
<pre style="display:block; text-align:left; border:1px solid #990000; border-radius:6px; padding:10px; background:#fff; margin:10px auto;">from whoosh import index
from whoosh.filedb.filestore import FileStorage

storage = FileStorage("/ruta/a/indice")
ix = storage.create_index(schema)</pre>
</li>
<li><code>RamStorage</code>:
<ul>
<li>Mantiene el índice completamente en memoria.</li>
<li>Muy rápido, pero volátil (se pierde al cerrar la aplicación).</li>
<li>Útil para pruebas o índices temporales.</li>
</ul>
<pre style="display:block; text-align:left; border:1px solid #990000; border-radius:6px; padding:10px; background:#fff; margin:10px auto;">from whoosh.filedb.filestore import RamStorage

storage = RamStorage()
ix = storage.create_index(schema)</pre>
</li>
</ul>
</li>
</ul>

<b>Paso 1: importar las librerías</b>

</div>


```python
from whoosh.qparser import *
from whoosh.filedb.filestore import RamStorage
from whoosh.fields import Schema, TEXT, KEYWORD, DATETIME, ID, STORED, NUMERIC
from whoosh.analysis import StemmingAnalyzer, StandardAnalyzer
from whoosh import index
from whoosh.writing import AsyncWriter, BufferedWriter
from datetime import datetime
import os, os.path
```

<div style="background-color:#333333; color:white; padding:15px; border-radius:8px; border:1px solid #000;">

<b>Paso 2: definir el schema</b>

</div>
<p>Cada documento puede tener múltiples campos, como título, contenido, URL, fecha, etc. Primero, necesitamos crear un esquema (schema) para nuestro corpus que especifique estos campos de los documentos en un índice.

El esquema es el conjunto de todos los campos posibles en un documento. Cada documento individual puede usar solo un subconjunto de los campos disponibles en el esquema.

Nota:
Sin un esquema, el query parser de Whoosh no procesará el texto en la consulta del usuario (es decir, no podrá hacer búsquedas por frases).
</p>

<p>
Por defecto, los campos de tipo <code>TEXT</code> no se almacenan, lo que significa que el contenido de este campo no se mostrará en el resultado de la búsqueda. Generalmente, no querrás almacenar el texto principal en el índice de búsqueda; sin embargo, puedes usar <code>TEXT(stored=True)</code> para especificar que el texto sí debe almacenarse en el índice.

El campo <code>KEYWORD</code> está diseñado para palabras clave separadas por espacios o comas. Este tipo de campo se indexa y es buscable (y opcionalmente almacenado). Ahora bien, no soporta búsquedas de frases.

Opciones importantes:
<ul>
    <li>Para almacenar el valor del campo en el índice, usa <code>stored=True</code> en el constructor.</li>
    <li>Para convertir automáticamente las palabras clave a minúsculas antes de indexarlas, usa <code>lowercase=True</code>. </li>
    <li>Para separar las palabras clave por comas (permitiendo que las palabras clave contengan espacios), usa <code>commas=True</code>.</li>
    <li>De lo contrario, las palabras clave se separan por espacios.</li>
</ul>

Para usar el campo keyword en búsquedas ponderadas, usa <code>scorable=True</code>.<br>

Aquí tienes varios ejemplos de cómo crear esquemas:
</p>


```python
# Schema básico
schema1 = Schema(title=TEXT(stored=True), content=TEXT)

# Schema con ID y fecha
schema2 = Schema(
    id=ID(stored=True), title=TEXT(stored=True), date=DATETIME(stored=True)
)

# Schema con temas y precio
schema3 = Schema(
    title=TEXT(stored=True),
    subject=KEYWORD(commas=True, scorable=True),
    price=NUMERIC(stored=True),
)
```

<div style="background-color:#333333; color:white; padding:15px; border-radius:8px; border:1px solid #000;">

<b>Paso 3: extraer la información e indexar los documentos</b>

</div>
<p>Después de crear nuestros esquemas, indexaremos cada documento en el corpus. Para ello haremos ejemplos siguiendo los tres esquemas.</p>
<p><b>Notas:</b><br>
<ol>
    <li>Los campos que se indexen deben recibir un valor unicode.</li>
    <li>Abrir un writer bloquea el índice para escritura. En un entorno multihilo o multiproceso, abrir un writer puede generar un error si ya hay un writer abierto. Probaremos cada esquema con un tipo de writer diferente.</li>
</ol>
</p>

<b>Caso del schema 1</b>

```python
# ----------------------------
# Schema 1 - IndexWriter (estándar)
# ----------------------------


# Crear carpeta si no existe
if not os.path.exists("index_schema1"):
    os.mkdir("index_schema1")

# Crear índice
ix1 = index.create_in("index_schema1", schema1)

# Abrir writer estándar para indexar documentos
writer1 = ix1.writer()  # IndexWriter

# Ejemplo de documentos
docs1 = [
    {"title": "Primer documento", "content": "Contenido de prueba"},
    {"title": "Segundo documento", "content": "Más contenido"},
]

for doc in docs1:
    # Nota: los valores deben ser unicode (str en Python 3)
    writer1.add_document(title=doc["title"], content=doc["content"])

writer1.commit()  # Cierra el writer y desbloquea el índice
print("Schema 1 indexado con IndexWriter")
```

<b>Caso del schema 2</b>

```python
# ----------------------------
# Schema 2 - AsyncWriter
# ----------------------------


# Crear carpeta si no existe
if not os.path.exists("index_schema2"):
    os.mkdir("index_schema2")

# Crear índice
ix2 = index.create_in("index_schema2", schema2)

# AsyncWriter permite escritura concurrente
writer2 = AsyncWriter(ix2)
docs2 = [
    {"id": "1", "title": "Libro A", "date": datetime(2020, 5, 1)},
    {"id": "2", "title": "Libro B", "date": datetime(2021, 7, 15)},
]

for doc in docs2:
    writer2.add_document(id=doc["id"], title=doc["title"], date=doc["date"])

writer2.commit()
print("Schema 2 indexado con AsyncWriter")
```

<b>Caso del schema 3</b>

```python
# ----------------------------
# Schema 3 - BufferedWriter
# ----------------------------

# Crear carpeta si no existe
if not os.path.exists("index_schema3"):
    os.mkdir("index_schema3")

# Crear índice
ix3 = index.create_in("index_schema3", schema3)

# BufferedWriter acumula cambios en memoria antes de escribir en disco
writer3 = BufferedWriter(
    ix3, period=10.0, limit=2
)  # period=tiempo máximo, limit=número máximo de docs en buffer
docs3 = [
    {"title": "Producto 1", "subject": "electrónica, televisión", "price": 19.99},
    {"title": "Producto 2", "subject": "textil, hombre", "price": 35.50},
]

for doc in docs3:
    writer3.add_document(title=doc["title"], subject=doc["subject"], price=doc["price"])

writer3.commit()
print("Schema 3 indexado con BufferedWriter")
```

<p>Después de indexar los documentos, podemos escribir la consulta y convertir la cadena de consulta en un objeto de consulta mediante el query parser.

Se crea un objeto <code>whoosh.qparser.QueryParser</code>, y luego se le pasa el nombre del campo por defecto sobre el cual se realizará la búsqueda y el schema del índice que se va a consultar.

De nuevo, probaremos consultas sobre cada esquema.

<b>Nota:</b>
¡Muy importante! La cadena de consulta debe ser un valor unicode.
</p>

<b>Caso del schema 1</b>

```python
from whoosh.qparser import QueryParser

with ix1.searcher() as searcher:
    # 1. Buscar por título
    q1 = QueryParser("title", ix1.schema).parse("Segundo")
    results1 = searcher.search(q1)
    print("Resultados con 'Segundo' en título:")
    for r in results1:
        print(r["title"])

    # 2. Buscar por contenido
    q2 = QueryParser("content", ix1.schema).parse("prueba")
    results2 = searcher.search(q2)
    print("\nResultados con 'prueba' en contenido:")
    for r in results2:
        print(r["title"])
```
